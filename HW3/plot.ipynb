```python
import pandas as pd
import matplotlib.pyplot as plt

PPO1=pd.read_csv('PPOP_1001.csv')
PPO2=pd.read_csv('PPOP_2001.csv')
PPO3=pd.read_csv('PPOP_3001.csv')
plt.plot(PPO1.iloc[:,2],label='seed=1001')
plt.plot(PPO2.iloc[:,2],label='seed=2001')
plt.plot(PPO3.iloc[:,2],label='seed=3001')
plt.title('PPO Pendulum-v0')
plt.legend()
```

```python
RFP1=pd.read_csv('RFPendulum-v0_1001.csv')
RFP2=pd.read_csv('RFPendulum-v0_2001.csv')
RFP3=pd.read_csv('RFPendulum-v0_3001.csv')
plt.plot(RFP1.iloc[:,2],label='seed=1001')
plt.plot(RFP2.iloc[:,2],label='seed=2001')
plt.plot(RFP3.iloc[:,2],label='seed=3001')
plt.title('RF Pendulum-v0')
plt.legend()
```

```python
plt.plot((PPO1.iloc[:,2]+PPO2.iloc[:,2]+PPO3.iloc[:,2])/3,label='PPO')
plt.plot((RFP1.iloc[:,2]+RFP2.iloc[:,2]+RFP3.iloc[:,2])/3,label='RF')
plt.legend()
plt.title('Pendulum-v0 RFvsPPO average')
```
